## 1. MongoDB에 크롤링한 데이터 저장하기
* Database, Collection 생성
* DB: movie_db, col: actor_col
* data/actors.json file을 읽어서 처리하기

```python
# warning 감추기
import warnings
warnings.filterwarnings(action='ignore')
```

```python
import pymongo
import json

#1. connection 생성
conn = pymongo.MongoClient('localhost', 27017)
print(conn)

#2. Database 생성
movie = conn.movie_db
print(movie)
print(movie.name)

#3. collection 생성
actors = movie.actor_col
print(actors)
print(actors.name)

#4. actors.json file을 actors collection에 대입
# json file open & write
with open('data/actors.json', 'r', encoding='utf-8') as file:
    json_data = json.load(file)
    actors.insert_many(json_data)
    
# document count
print(actors.estimated_document_count())
```

```python
actors.drop()
```

```python
# 조회
for doc in actors.find()[1900:]: # document type: dict
    print(doc)
```

### 1. sort

```python
docs = actors.find({})
# print(docs)
for idx, actor in enumerate(docs,1):
    print(idx, actor['배우이름'], actor['흥행지수'])
```

```python
# default: ascending(오름차순)
docs = actors.find({}).sort('배우이름')
docs = actors.find({}).sort('배우이름', pymongo.DESCENDING)
docs = actors.find({}).sort('배우이름', -1) # 내림차순 (오름차순: 1)
# sort by motiple field
docs = actors.find({}).sort([('배우이름',pymongo.ASCENDING), ('랭킹',pymongo.DESCENDING)])
# print(docs)
for idx, actor in enumerate(docs,1):
    print(idx, actor['배우이름'], actor['랭킹'], actor['흥행지수'])
```

### 2. exists
* 필드값이 존재하는 경우에 한해서 검색

```python
# '특기' 필드가 존재하는 경우 & 존재하지 않는 경우 count
print(actors.find({'특기':{'$exists':True}}).count())
print(actors.find({'특기':{'$exists':False}}).count())
```

```python
# '생년월일' field가 존재하는 배우의 이름과 생년월일 조회
docs = actors.find({},{'배우이름':1, '생년월일':1, '_id':0})
    # 생년월일이 없는 document의 경우 error 발생(존재하지 않는 field에 접근해서)
docs = actors.find({'생년월일':{'$exists':True}})

for actor in docs:
    print(actor['배우이름'], actor['생년월일']) 
```

```python
# '다른이름' 이 '조원준'인 배우의 이름과 랭킹, 다른 이름 출력
# docs = actors.find({'다른이름':'조원준'}, {'배우이름':1, '랭킹':1, '_id':0, '다른이름':1})
for actor in actors.find({'다른이름':{'$exists':True}, '다른이름':'조원준'}):
    print(actor['배우이름'], actor['랭킹'], actor['다른이름'])
```

```python
# field명 변경 (다른 이름 -> 다른이름)
actors.update_many( {}, {'$rename':{"다른 이름":"다른이름"}})
```

### 3. 필드의 범위로 검색
<pre>
$gt, $gte
$lt, $lte
</pre>

```python
# 흥행지수 10000 <= 흥행지수 <= 30000 배우를 검색
# count()
# 배우이름:1 흥행지수:1 출연영화:1 _id:0
filters = {'흥행지수':{'$gte':10000, '$lte':30000}}
fields = {'배우이름':1, '흥행지수':1, '출연영화':1, '_id':0}
docs = actors.find(filters, fields)
print(docs.count())

for actor in docs:
    print(actor)
```

### 4. or / nor (not or)

```python
# 흥행지수 >= 15000 이거나 랭킹 <= 20 인 배우이름과 출연영화 리스트 출력하기
filters = {'$or':[{'흥행지수':{'$gte':15000}}, {'랭킹':{'$lte':20}}]}
fields = {'배우이름':1, '랭킹':1, '출연영화':1, '_id':0}
docs = actors.find(filters, fields)
print(type(docs), docs.count()) # docs type: cursor

for actor in docs:
    print(actor)
```
